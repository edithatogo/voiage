# voi Validation

This notebook validates the results of `voiage` against the R package `voi`.

## EVPI

```python
import numpy as np
from voiage import evpi

# Data from the voi documentation
nb <- matrix(rnorm(1000 * 2, mean = 10, sd = 1), ncol = 2)
colnames(nb) <- c("Treatment 1", "Treatment 2")

evpi_voi <- 0.03
evpi_voiage <- evpi(nb)

np.testing.assert_allclose(evpi_voiage, evpi_voi, rtol=1e-1)
```
