# EVPI Validation Notebook

A compact, runnable EVPI check using a deterministic two-strategy example.

```
import numpy as np
from voiage.analysis import DecisionAnalysis
from voiage.schema import ValueArray, ParameterSet, TrialDesign, DecisionOption
from voiage.methods.sample_information import evsi, enbs

np.random.seed(7)
values = np.column_stack([
    np.full(100, 10.0) + np.random.normal(0.0, 0.4, 100),
    np.full(100, 11.0) + np.random.normal(0.0, 0.4, 100),
])
value_array = ValueArray.from_numpy(values, ['standard care', 'new treatment'])
analysis = DecisionAnalysis(nb_array=value_array)
evpi_value = analysis.evpi()
print(f'EVPI: {evpi_value:.4f}')
```

```
theoretical_evpi = float(np.mean(np.max(values, axis=1)) - np.max(np.mean(values, axis=0)))
print(f'Theoretical EVPI: {theoretical_evpi:.4f}')
```
