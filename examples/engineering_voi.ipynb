# Engineering VOI

This notebook shows a compact engineering decision problem with three design options. We compare expected value, EVPI, and EVPPI for a small reliability model.

```
import numpy as np
from voiage.schema import ValueArray
from voiage.methods import evpi, evppi

np.random.seed(7)
```

```
samples = 240
designs = ["Baseline", "Reinforced", "Monitored"]

# Synthetic engineering net benefit: production value minus expected rework/failure cost.
baseline_cost = np.random.normal(15.0, 2.0, size=samples)
reinforced_cost = np.random.normal(12.0, 1.5, size=samples)
monitored_cost = np.random.normal(13.0, 1.2, size=samples)

baseline_value = np.random.normal(25.0, 1.0, size=samples)
reinforced_value = np.random.normal(24.5, 0.8, size=samples)
monitored_value = np.random.normal(24.8, 0.9, size=samples)

net_benefit = np.column_stack([
    baseline_value - baseline_cost,
    reinforced_value - reinforced_cost,
    monitored_value - monitored_cost,
])

value_array = ValueArray.from_numpy(net_benefit, strategy_names=designs)
value_array
```

```
expected_net_benefit = net_benefit.mean(axis=0)
dict(zip(designs, expected_net_benefit.round(3), strict=True))
```

```
parameter_samples = {
    "failure_rate": np.random.rand(samples),
    "inspection_cost": np.random.rand(samples),
}

engineering_evpi = evpi(value_array)
engineering_evppi = evppi(
    net_benefit,
    parameter_samples=parameter_samples,
    parameters_of_interest=["failure_rate", "inspection_cost"],
)

print(f"Engineering EVPI: {engineering_evpi:.4f}")
print(f"Engineering EVPPI: {engineering_evppi:.4f}")
```

A positive EVPI indicates that better reliability information still has decision value. The EVPPI slice is intentionally synthetic here; the point is to show the engineering workflow with the public VOI API rather than reproduce a particular published case study.
