# EVPPI Validation Notebook

A compact, runnable EVPPI check using a small parameter set and the stable API.

```
import numpy as np
from voiage.analysis import DecisionAnalysis
from voiage.schema import ValueArray, ParameterSet, TrialDesign, DecisionOption
from voiage.methods.sample_information import evsi, enbs

np.random.seed(8)
values = np.column_stack([
    np.full(120, 9.5) + np.random.normal(0.0, 0.5, 120),
    np.full(120, 10.8) + np.random.normal(0.0, 0.5, 120),
])
value_array = ValueArray.from_numpy(values, ['standard care', 'new treatment'])
parameters = ParameterSet.from_numpy_or_dict({
    'treatment_effect': np.linspace(0.1, 0.9, 120),
    'cost_shift': np.linspace(-0.2, 0.2, 120),
})
analysis = DecisionAnalysis(nb_array=value_array, parameter_samples=parameters)
evppi_value = analysis.evppi(n_regression_samples=50)
print(f'EVPPI: {evppi_value:.4f}')
```

```
evpi_value = analysis.evpi()
print(f'EVPI upper bound: {evpi_value:.4f}')
```
